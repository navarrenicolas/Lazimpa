```python
def successors(n):
    succs = []
    count = 0
    for message in messages:
        locs = np.where(message==n)[0]
        if len(locs) > 0:
            count += 1
            for idx in locs:
                if idx+1 < len(message):
                    succs.append(message[idx+1])
                else: 
                    succs.append(-1)
    return (count,succs)
```

```python
def informative_len(msg):
    return len(msg[msg==0])
```

```python
def message_length(msg):
    snip = lambda x: x[x>=0]
    return len(snip(msg))
```

```python
import random
def shuffle(x):
    return random.sample(x, len(x)) 
```

### Successive pair map

```python
bigram_transitions = np.array([[successors(i)[1].count(j) for i in range(40)] for j in np.flip(np.arange(-1,40))])
```

```python
ax = heatmap(np.log(bigram_transitions+1))
ax.invert_yaxis()
```

```python
neighb_trans = 0.0
out_trans = 0.0
eos_trans = 0.0
self_trans = 0.0
tot_tansitions = 0

for word in range(40):
    word_trans = bigram_transitions[:,word]
    tot_tansitions += sum(word_trans)
    for vocab, transitions in enumerate(word_trans):
        if vocab == 40:
            eos_trans += transitions
            break
        if vocab in neighbs[word]:
            neighb_trans += transitions
        else:
            if vocab == word:
                self_trans += transitions
            else:
                out_trans += transitions
neighb_pct = neighb_trans/tot_tansitions
out_pct = out_trans/tot_tansitions
self_pct = self_trans/tot_tansitions
eos_pct = eos_trans/tot_tansitions
print(out_pct+eos_pct+neighb_pct+self_pct)
```

```python
print(f'Neighbs: {neighb_pct*100}, Out: {100*out_pct}, Repeat: {100*self_pct}, EOS: {100*eos_pct}') 
```

```python
N_perms = int(1e5)
counts = 0.0
mwm = mean_wm(messages,power_law)
for i in range(N_perms):
    if mwm >= mean_wm(messages,shuffle(power_law)):
        counts = counts + 1
p_val = counts/(N_perms*1.0)
print('%.9f' % p_val)
```
